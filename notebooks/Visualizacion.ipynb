```python
## Diagrama de dispersion (scatterpol)

x= np.arange(2,20)

y= x**2

y2=x**0.5

plt.scatter(x , y , color='green',  marker='*' , label='x a la 2 ')
plt.scatter(x , y2 , color='black',  marker='*' , label='x a la 0.5')
plt.title('Funcion de x fx(x)', fontsize=16 , color='red')


## Otro grafico 
plt.plot(x,y2 , color='pink')


plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()


plt.text(5,100,'TEXTO', color='blue')
```
